```python
import re
def clean_author(author):
    regex_pattern = r"'([^']+)'"
    extracted_values = re.findall(regex_pattern, author)
    match = re.search(regex_pattern, author)
    return extracted_values if extracted_values else ''
```

```python
df['authors'] = df['authors'].apply(lambda authors: clean_author(authors))

df
```

```python
def insert_research(engine, df):
    # convert authors into json format for dumps into database
    

    df.to_sql('research', con=engine, if_exists='append', index=False)
    print("insert data successfully for :", len(df), "records")
```
